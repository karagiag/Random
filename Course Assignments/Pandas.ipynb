```python
# Create data
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35], 
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9], 
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
```

```python
# Create Data Frame
df = pd.DataFrame(raw_data)

#Set index
df = df.set_index(df.origin)
```

```python
df.head()
```

```python
#Select multiple columns
df[['battles','company']]
```

```python
# Select all rows with the index label "Arizona"
df.loc[:'Arizona']
```

```python
# Another way
df[df.index.isin(['Arizona'])]
```

```python
# Select all rows with the index label is not "Arizona"
df[~df.index.isin(['Arizona'])]
```

```python
# Select every row up to 3
df.iloc[:2]
```

```python
# Select the second and third row
df.iloc[1:3]
```

```python
# Select every row beginning with the third row
df.iloc[2:]
```

```python
# Select the first 2 columns
df.iloc[:,:2]
```

```python
# Select the second and thrird columns of the fist 3 rows
df.iloc[:3,1:3]
```
