```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C:/Users/karagiag/Downloads/aapl.csv', index_col=0, parse_dates=True)
```

```python
data = data.sort_index(0)
```

```python
data.rename(columns={'Close':'price'},inplace=True)
```

```python
data
```

```python
data['returns']=np.log(data['price']/data['price'].shift(1))
```

```python
data.head()
```

```python
data['position']=np.sign(data['returns'])
```

```python
data['strategy']=data['position'].shift(1)* data['returns']
```

```python
data[['returns', 'strategy']].dropna().cumsum().apply(np.exp).plot(figsize=(10, 6))
```

```python
plt.show()
```

```python
# Instead of using the sign of the last return as indication of momentum,
# we could use the sign of the average of the last two returns as a signal for positioning

data['position'] = np.sign(data['returns'].rolling(6).mean())  
```

```python
data['strategy'] = data['position'].shift(1)*data['returns']
```

```python
data[['returns', 'strategy']].dropna().cumsum().apply(np.exp).plot(figsize=(10, 6))
```

```python
plt.show()
```
