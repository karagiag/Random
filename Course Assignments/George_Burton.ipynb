```python
good_data = []

for file_name in glob.glob(os.path.join(input_path,'*.csv')):
    f = open(file_name)
    lines = f.readlines()
    data = pd.DataFrame(map(lambda x: x.replace('\n', '').split(','), lines[4:]), columns=lines[3].replace('\n', '').split(','))
    
    participant_id = file_name.split('\\')[-1].split('_')[1]
    
    d = data.convert_objects(convert_numeric=True)
    d['participant_id'] = participant_id
    
    condition1 = d[d.correct == 1].RT.mean() + 3 * d[d.correct == 1].RT.std()
    condition2 = d[d.correct == 1].RT.mean() - 3 * d[d.correct == 1].RT.std()
    
    good_data.append(d[(d.RT < condition1) & (d.RT > condition2) & (d.correct == 1)])

good_data = pd.concat(good_data)
```

```python
good_data.pivot_table(index=['participant_id'], columns=['condition2'], values='RT', aggfunc='mean').to_clipboard()
```
